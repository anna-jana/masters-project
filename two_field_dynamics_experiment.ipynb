```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.integrate as si
```

```python
def rhs(t, y, m_a, m_chi, g):
    H = 1 / (2*t)
    a, a_dot, chi, chi_dot = y
    return (
        a_dot,
        - 3 * H * a_dot - m_a**2 * a - g * a * chi**2,
        chi_dot,
        - 3 * H * a_dot - m_a**2 * a - g * a * chi**2,
    )
```

```python
a0 = 1e10
chi0 = 1e10
H_inf = 1800000.0
t0 = 1 / H_inf
sol = si.solve_ivp(rhs, (t0, 1e5 * t0), (a0, 0.0, chi0, 0.0), args=(1e-3, 1e-3, 1))
plt.semilogx(sol.t, sol.y[0] / a0, label="a / a0")
plt.semilogx(sol.t, sol.y[2] / chi0, label="chi / chi0")
plt.legend()
plt.show()
plt.semilogx(sol.t, sol.y[0] / a0 - sol.y[2] / chi0)
```
